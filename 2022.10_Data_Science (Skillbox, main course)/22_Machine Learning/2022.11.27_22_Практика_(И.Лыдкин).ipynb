1. Подготовка базовой модели

Обучите простую модель классификации с помощью DecisionTreeClassifier на данных из датасета vehicles_dataset_prepared.csv. Для этого сделайте шаги:

1. Обучите модель дерева решений с зафиксированным random_state на тренировочной выборке.
2. Сделайте предикт на тестовой выборке.
3. Замерьте точность на тестовой выборке и выведите матрицу ошибок. 
4. Удалите фичи с нулевыми весами по feature_importance из тренировочной и тестовой выборок.
5. Заново обучите модель и измерьте качество.

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score, confusion_matrix 
````

````python
df = pd.read_csv('data/vehicles_dataset_prepared.csv')

df_prepared = df.copy()
df_prepared = df_prepared.drop(['price', 'odometer/price_std'], axis=1)

x = df_prepared.drop(['price_category'], axis=1)
y = df_prepared['price_category']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
df.shape
````

````python
clf = DecisionTreeClassifier()
````

````python
clf.fit (x_train, y_train)
````

````python
predicted_test = clf.predict(x_test)
````

````python
print (accuracy_score (y_test, predicted_test))
````

````python
confusion_matrix(y_test, predicted_test)
````

````python
f_imp_list = list (zip (x_train.columns, clf.feature_importances_))
f_imp_list.sort (key = lambda x: x[1], reverse = True)
f_imp_list
````

````python
relevant_features = []
for i in range (len (f_imp_list)):
    if f_imp_list[i][1] != 0:
        relevant_features.append(f_imp_list[i][0])
relevant_features
````

````python
df_prepared2 = df_prepared[relevant_features]
df_prepared2.shape
````

````python
x1 = df_prepared2
y1 = df_prepared['price_category']
````

````python
x1_train, x1_test, y1_train, y1_test = train_test_split (x1, y1, test_size = 0.3, random_state = 42)
````

````python
clf.fit (x1_train, y1_train)
````

````python
predicted_test1 = clf.predict(x1_test)
````

````python
print (accuracy_score (y1_test, predicted_test1))
````

````python
confusion_matrix(y_test, predicted_test)
````

2. Подготовка модели случайного леса

Обучите простую модель классификации с помощью RandomForestClassifier. Для этого на новых урезанных семплах тренировочной и тестовой выборок обучите модель случайного леса с зафиксированным random_state=50. 

````python
from sklearn.ensemble import RandomForestClassifier
x1_train, x1_test, y1_train, y1_test = train_test_split (x1, y1, test_size = 0.3, random_state = 50)
rf_clf = RandomForestClassifier()
rf_clf.fit(x1_train, y1_train)
````

2. Сделайте предикт, посчитайте точность модели и матрицу ошибок. Сравните с предыдущей моделью дерева решений. Есть ли случаи, когда модель из пункта 1 отрабатывает лучше, чем модель случайного леса?

**Если каждое "дерево" в "лесу" обладает точностью выше 50%, то кличество ошибившихся деревьев в таком лесу будет с высокой вероятностью меньше половины. Таким образом, метод голосования простым большинством значительно снижает вероятность ошибки, т.е. повышает эффективность предсказания (пока не могу выразить эту мысль математически). Т.о. предсказательная способность леса должна быть всегда выше, чем у одно дерева.**

**При этом дерево менее ресурсоемко и нагляднее, что упрощает калибровку под конкретную задачу**

**Предполагаю, что на практике на 1 дереве отрабатывают идеи по подготовке данных, а затем запускают "лес"**

````python
predicted_train_rf = rf_clf.predict(x1_train)
predicted_test_rf = rf_clf.predict(x1_test)
print(accuracy_score(y1_train, predicted_train_rf))
print(accuracy_score(y1_test, predicted_test_rf))
````

````python
confusion_matrix(y1_test, predicted_test_rf)
````
