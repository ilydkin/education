```python
def get_month(x):
    return dt.datetime(x.year,x.month,1)
```

```python
data['month_year']=data['date'].apply(get_month)
```

```python
data.head()
```

```python
data['cohortMonth']=data.groupby('user_id')['month_year'].transform('min')
```

```python
data.head(1)
```

```python
def get_date(df,column):
    year=df[column].dt.year
    month=df[column].dt.month
    day=df[column].dt.day
    return year,month,day
```

```python
date_year,date_month,_=get_date(data,'date')
cohort_year,cohort_month,_=get_date(data,'cohortMonth')

year_diff=date_year-cohort_year
month_diff=date_month-cohort_month

data['cohort_index']=year_diff*12+month_diff+1
data.head(1)
```

```python
#сумма покупок в штуках
order_n = data.pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='user_id',
                      aggfunc=len)

first_column_o=order_n.iloc[:,0]
retention_o=order_n.div(first_column_o,axis=0)

plt.figure(figsize=(11,9))
plt.title('ORDERS \n когортный анализ-retention')
sns.heatmap(data=retention_o,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

```python
#количество уникальных клиентов в месяц по когортам

data['cohort_nunique']=data.groupby(['cohort_index','cohortMonth'])['user_id'].transform('nunique')

unique_c = data.pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='cohort_nunique')

first_column_u=unique_c.iloc[:,0]
retention_u=unique_c.div(first_column_u,axis=0)

plt.figure(figsize=(11,9))
plt.title('UNIQUE CUSTOMERS \n когортный анализ-retention')
sns.heatmap(data=retention_u,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```
