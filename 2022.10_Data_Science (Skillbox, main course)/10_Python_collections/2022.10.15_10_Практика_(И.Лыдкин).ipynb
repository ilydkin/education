```python
# Задача 4.1 Количество уникальных катергорий
unique_len = print(f'Количество уникальных катергорий: {len(set([i[:4] for i in items_count.keys()]))}')
```

```python
# Задача 4.2 Среднее количество единиц в категории 
def sum_per_cat (items_count):
    new_dict = {}
    for k, v in items_count.items():
        if k[:4] not in new_dict:
            new_dict[k[:4]] = 0
        new_dict[k[:4]] += v
    return new_dict
```

```python
b = sum_per_cat (items_count)
print (f'Среднее количество единиц в категории: {round (sum (b.values()) / len (b.values()), 1)}')
```

```python
# Задача 4.3 Среднее количество уникальных товаров в категории
def cat_frequency (items_count):
    new_dict = {}
    for i in sum_per_cat (items_count).keys():
        new_dict[i] = 0
        for elem in items_count.keys():         
            if i == elem[:4]:
                new_dict[i] += 1
    return new_dict
```

```python
c = cat_frequency (items_count)
print (f'Среднее количество уникальных товаров в категории: {(sum (c.values()) / len (c.values()))}')
```

В переменной `document_text` задан текст. Нужно найти частоту использования каждого уникального слова и положить его в словарь `word_frequency`. Преобразовывать слова в начальную форму не нужно.

Рекомендации:

2. Для подсчёта частот используйте метод `count` на списках.


```python
document_text = 'There are the Food Courts in the cities that seats a lot of people and caters for every taste with dishes from all around the world. \
 The aromatic smells that come from the Food Courts arc so delicious that your mouth will water. \
 In recent years foreign foods have become a regular part of our life.'
```

```python
#Задача 5 Контент анализ текста
def normalization (document_text): 
    return document_text.replace('.', '').replace('  ',' ').split(" ")
# Я здесь сначала еще регистр понизил (.lower()), но тогда с ответом не сходится.
# Хотя, выровнять регистры было бы правильнее, я думаю.
```

```python
def frequency (document_text):
    dictionary = {}
    normal_text = normalization (document_text)
    for i in normal_text:
        if i not in dictionary:
            dictionary[i] = normal_text.count(i) 
    return (dictionary)
```

```python
## Здесь сделал сортировку до того, как прочитал следующее задание. Решил оставить
# Так же можно было бы загнать значения в новый словарь и вывести на печать его (чтоб совпало по оформлению с ответом),
# Но ведь так симпатичнее). С учетом выровненного регистра вообще было красиво.

word_frequency = frequency (document_text)
for k, v in sorted (word_frequency.items()): 
    print (f'{k} : {v}')
```

```python
#Задача 6. Частота слова. Слова из word_frequency в порядке убывания частоты.
#Если частота одинаковая - слова в обратном лексикографическом порядке

def reverted_dict (word_frequency):
    new_dict = {}
    
    for k, v in word_frequency.items():
        if str(v) not in new_dict:
            new_dict[str(v)] = []
        new_dict[str(v)].append(k)
    
    return new_dict
```

```python
def arrangement (word_frequency):
    r_dict = reverted_dict (word_frequency)
    arranged_dict = dict(sorted (r_dict.items(), reverse = True))
    sorted_list = []
    
    for i in arranged_dict:
        for j in sorted (arranged_dict[i], reverse = True):
            sorted_list.append (j)
    
    return sorted_list
```
