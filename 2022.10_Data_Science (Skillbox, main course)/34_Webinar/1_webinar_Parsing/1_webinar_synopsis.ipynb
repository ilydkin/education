```python
# Теперь нужно вытащить все, похожее на "/rus_text_pesni/46766-grigorijj-leps-odin-fig.html"
songs_lists = []

for a in songs_soup.find_all('a'):
    if str(a).find('rus_text_pesni') != -1 and str(a).find('leps') != -1:
        song_url = "https://text-you.ru/" + str(a['href'])
        songs_lists.append(song_url)
songs_lists
#Это ссылки на все песьни на 1 странице. Страниц 3.
```

```python
# Итерируем по 3 страницам

urls = ['https://text-you.ru/songs/григорий+лепс.html', 'https://text-you.ru/songs/григорий+лепс-2.html',
       'https://text-you.ru/songs/григорий+лепс-3.html']

songs_lists = []

for url in urls:
    songs_list = requests.get(url)
    
    songs_soup = BeautifulSoup(songs_list.content, "html.parser")    
    for a in songs_soup.find_all('a'):
        if str(a).find('rus_text_pesni') != -1 and str(a).find('leps') != -1:
            song_url = "https://text-you.ru/" + str(a['href'])
            songs_lists.append(song_url)
            
songs_lists = list(set(songs_lists)) #убивает дубликаты
print (len(songs_lists))
```

```python
# Вытаскиваем текст по ссылке
# Проверка связи
song_name = songs_lists[0] #берем 1 ссылку на пробу

song_url = requests.get(song_name)
song_url.status_code
```

```python
#парсим возврат по ссылке, вытаскиваем текст
#по F12 видим, что текст идет после <pre>
song_soup = BeautifulSoup(song_url.content, 'html.parser')
song_text = str(song_soup.find_all('pre')[0]) #[0] вытаскивает содержимое из списка. Без него текст выводится в []
song_text
```

```python
# пишем функцию нормализации текста
def replace_all(text, to_replace):
    for i in to_replace:
        text = text.replace(i, ' ')
    return text
```

```python
replace_all (song_text, ['<pre>', '<br/>', '</pre>'])
```

```python
song_text = replace_all (song_text, ['<pre>', '<br/>', '</pre>'])
song_text = song_text.lstrip().rstrip() #убирает пробелы слева и справа соответственно
song_text
```

```python
parsed_songs = {} # словарик "url" :'текст песни'
n = 0 #счетчик


for song_name in songs_lists:
    song_url = requests.get(song_name)
    song_soup = BeautifulSoup(song_url.content, "html.parser")
    song_text = str(song_soup.find_all('pre')[0])
    song_text = replace_all(song_text, ['<pre>', '<br/>', '</pre>'])
    song_text = song_text.rstrip().lstrip()
    parsed_songs[song_name] = song_text
    n += 1
    print('Parsed', n, '/', len(songs_lists))
    time.sleep(1) # Это для того, чтобы не заблокировал сайт 
```
