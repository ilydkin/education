# 2. Add Two Numbers (linked lists)
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list. You may assume the two numbers do not contain any leading zero, except the number 0 itself.

```python
# My code which is almost equivalent to the examplary
class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1, l2):

        head = Node()
        current = head
        carry = 0
        
        while l1 or l2 or carry:
            x = l1.val if l1 else 0
            y = l2.val if l2 else 0
            sum_ = x + y + carry
            carry = sum_ // 10
            
            new = Node(sum_ % 10)
            current.next = new
            
            current = current.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        
        return head.next  
```

### Some linked lists theory
It was difficult to understand why the `head` is needed, why it is linked with `current`, and how, while working with `current` in a loop, we can return `head.next`.

Explanation:

1. The `head` serves as the starting point of the linked list. It is the first `Node` with `val = 0`, and `.next` refering to the `Node.val` of the first actual element. It allows us to "unwind" (display) the entire list. Initially, `head.next` is empty.

2. Initially, both `head` and `current` point to the same object - the dummy `Node (val=0, next=None)`. During the first iteration of the loop, `current.next` is assigned the value of the first real Node (`current.next = new`). Then, in each iteration, `current` is reassigned to the next element in the list.

3. The trick with displaying of smth, which appears as a list, is achieved by the LeetCode interpreter's __str__ function, which is not shown and most likely looks like this:

$$def __str__(self):
    node = self.head
    line = '['
    while node.next:
        line += str(node.val) + ','
        node = node.next
    line += str(node.val) + ']'
    return line$$

Therefore, by passing `head.next` to this function, we simply provide the starting Node from which it "unwinds" the list.
